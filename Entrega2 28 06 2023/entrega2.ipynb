```python
# importo las librerias a usar
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# leo el archivo, y pongo la columna 0 como indice
df=pd.read_csv("./afa_2015_2022_spa.csv", index_col=0)
df
```

```python
df_goles_por_campeonato = df.groupby('torneo')['goles_local','goles_visitante'].sum()

# Obtener los datos de goles locales y visitantes
goles_locales = df_goles_por_campeonato['goles_local']
goles_visitantes = df_goles_por_campeonato['goles_visitante']

# Crear una lista con los nombres de los campeonatos
campeonatos = df_goles_por_campeonato.index.tolist()

# Configurar el grÃ¡fico de barras
plt.bar(campeonatos, goles_locales, label='Goles Locales')
plt.bar(campeonatos, goles_visitantes, label='Goles Visitantes')
plt.xticks(rotation=45)


# Agregar etiquetas y tÃ­tulo
plt.xlabel('Campeonato')
plt.ylabel('Cantidad de Goles')
plt.title('Goles Locales y Visitantes por Campeonato')

# Mostrar la leyenda
plt.legend()

# Mostrar el grÃ¡fico
plt.show()
```
